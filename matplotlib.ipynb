```python
import matplotlib.pyplot as plt
import numpy as np

monthNumber = np.arange(1, 13)
profitList = np.array([211000, 183300, 210000, 210000, 200000, 198000,
                       290000, 385000, 245000, 270000, 412000, 300000])
plt.plot(
    monthNumber,
    profitList,
    linestyle='--',
    color='red',
    linewidth=3,
    marker='o',
    markerfacecolor='black',
    markeredgecolor='red',
    label='Profit data of last year'
)


plt.title("Company Sales data of last year")
plt.xlabel("Month Number")
plt.ylabel("Profit in dollar")


plt.legend(loc='lower right')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


monthNumber = np.arange(1, 13)
#salesunitsinnumber = np.array(1000,18000)


facecream_sales =  [2500, 2630, 2140, 3400, 3600, 2760, 2980, 3700, 3540, 1990, 2340, 2900]
facewash_sales =   [1500, 1200, 1340, 1130, 1740, 1555, 1680, 1900, 1600, 1700, 2100, 2200]
toothpaste_sales = [5200, 5100, 4550, 5870, 4560, 4890, 4780, 6100, 6300, 6250, 6000, 6200]
bathingsoap_sales = [9200, 9550, 8760, 8900, 8960, 9300, 9550, 9650, 9300, 9100, 9250, 9700]
shampoo_sales =    [1200, 2100, 1340, 1130, 1740, 1555, 1680, 1900, 1600, 1700, 2100, 2200]
moisturizer_sales = [1500, 1200, 1340, 1130, 1740, 1555, 1680, 1900, 1600, 1700, 2100, 2200]

plt.figure(figsize=(10,6))

plt.plot(monthNumber, facecream_sales,  marker='o', label='Face cream Sales Data')
plt.plot(monthNumber, facewash_sales,   marker='o', label='Face Wash Sales Data')
plt.plot(monthNumber, toothpaste_sales, marker='o', label='ToothPaste Sales Data')
plt.plot(monthNumber, bathingsoap_sales,marker='o', label='Bathing Soap Sales Data')
plt.plot(monthNumber, shampoo_sales,    marker='o', label='Shampoo Sales Data')
plt.plot(monthNumber, moisturizer_sales,marker='o', label='Moisturizer Sales Data')


plt.title("Sales data")
plt.xlabel("Month Number")
plt.ylabel("Sales units in number")
plt.legend(loc='upper left')
plt.grid(True)


plt.show()
```
